```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram of alcohol-related accidents
plt.figure(figsize=(8, 6))
sns.histplot(df['alcohol'], bins=20, kde=True)
plt.xlabel('Alcohol-Related Accidents')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol-Related Accidents')
plt.show()
```

Inference - The histogram shows that the majority of accidents have a low to average alcohol-related factor.

```python
sns.scatterplot(x="speeding",y="total",data=df)
```

Inference - we can observe that with increase in speed, there is increase in total accidents

```python
sns.relplot(x="speeding",y="total",data=df,hue="alcohol")
```

Inference - Speeding & Total accidents & Alcoholic drivers are directly proportional

```python
sns.histplot(df["total"])
```

Inference - distribution of total crashes

```python
sns.boxplot(x="abbrev", y="ins_premium", data=df)
plt.xlabel("State Abbreviation")
plt.ylabel("Insurance Premiums")
plt.xticks(rotation=90)
plt.show()
```

Inference -  variation in insurance premiums across different states.

```python
sns.barplot(x="abbrev", y="not_distracted", data=df,errorbar=None)
plt.xlabel("State Abbreviation")
plt.ylabel("Not Distracted Crashes")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="no_previous", y="ins_premium", data=df)
plt.xlabel("No Previous Crashes")
plt.ylabel("Insurance Premiums")
plt.title("Scatterplot of No Previous Crashes vs. Insurance Premiums")
plt.show()
```

Inference - the number of no previous crashes is not related to insurance premiums

```python
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```

Inference - ins_premium and ins_losses are not much correlated with other columns

```python
sns.pairplot(df)
plt.show()
```
